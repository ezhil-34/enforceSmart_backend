```python
import pandas as pd
import numpy as np
import joblib
from fastapi import FastAPI

# -------------------------------
# LOAD MODELS
# -------------------------------
dbscan = joblib.load("model/dbscan.pkl")
scaler = joblib.load("model/scaler.pkl")

# -------------------------------
# CONFIG
# -------------------------------
DATA_PATH = "jan to may police violation_anonymized791b166.csv"

app = FastAPI(title="EnforceSmart AI – Parking Hotspot Prediction")

# -------------------------------
# LOAD & PREPROCESS DATA
# -------------------------------
df = pd.read_csv(DATA_PATH)

df["created_datetime"] = pd.to_datetime(
    df["created_datetime"],
    utc=True,
    format="mixed",
    errors="coerce"
)

df = df.dropna(subset=["created_datetime"])
df["hour"] = df["created_datetime"].dt.hour

# -------------------------------
# CLUSTERING
# -------------------------------
coords = np.radians(df[["latitude", "longitude"]].to_numpy())
df["cluster"] = dbscan.fit_predict(coords)
df = df[df["cluster"] != -1]

# -------------------------------
# FEATURE ENGINEERING
# -------------------------------
vehicle_weight = {
    "SCOOTER": 0.3,
    "CAR": 0.6,
    "MAXI-CAB": 0.9,
    "LCV": 0.8
}

df["vehicle_weight"] = df["vehicle_type"].map(vehicle_weight).fillna(0.5)

def violation_score(v):
    v = str(v)
    if "MAIN ROAD" in v or "ROAD CROSSING" in v:
        return 1.0
    if "NO PARKING" in v:
        return 0.7
    return 0.5

df["violation_severity"] = df["violation_type"].apply(violation_score)

# -------------------------------
# HOTSPOT PREDICTION
# -------------------------------
hotspots = []

for cid, g in df.groupby("cluster"):
    raw_score = (
        0.45 * len(g) +
        0.30 * g["vehicle_weight"].mean() * 100 +
        0.25 * g["violation_severity"].mean() * 100
    )

    score = scaler.transform(
        pd.DataFrame([[raw_score]], columns=["raw_score"])
    )[0][0]

    band = "high" if score >= 70 else "medium" if score >= 40 else "low"

    peak_hour = g["hour"].value_counts().idxmax()
    peak_window = f"{peak_hour:02d}:00–{(peak_hour+3)%24:02d}:00"

    hotspots.append({
        "id": f"H-{int(cid)}",
        "latitude": g["latitude"].mean(),
        "longitude": g["longitude"].mean(),
        "score": int(score),
        "band": band,
        "peak": peak_window,
        "violationFrequency": int(len(g)),
        "dominantVehicle": g["vehicle_type"].mode()[0],
        "dominantViolation": g["violation_type"].mode()[0]
    })

# -------------------------------
# API ENDPOINT
# -------------------------------
@app.get("/predict/hotspots")
def predict_hotspots():
    return {
        "total_hotspots": len(hotspots),
        "hotspots": hotspots
    }
```
